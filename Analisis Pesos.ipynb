```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def graficarPorcentajes(pesos, titulos):
    dic = {}
    for i in range(len(titulos)):
        porcentajes_pesos = []
        for j in range(len(pesos[i])):
            porcentajes_pesos.append(pesos[i][j]/sum(pesos[i])*100)
        
        dic[titulos[i]] = porcentajes_pesos
        
    df_pesos = pd.DataFrame(dic).T
    df_pesos.columns=['XBoost basado en Benchmark', 'XGBoost basado en KNN', 'Catboost', 'KNN']

    g = df_pesos.plot(kind='bar', stacked=False, rot=0, figsize=(10,6))
    g.grid(axis='y')
    
    g.set_title('Porcentaje del peso asignado a cada modelo\n seg√∫n su Score',
               fontsize=20)
    g.set_ylabel('Porcentaje del peso(%)', fontsize=16)
    plt.legend(loc='lower center')
```

```python
graficarPorcentajes([[4,3,2,1],[4,3,2,0]], ["Ranking", "Ranking sin KNN"])
```

```python
graficarPorcentajes([[469,460,395, 348],[469,460,395, 145.64]], ['Score','Score reduciendo KNN'])
```
