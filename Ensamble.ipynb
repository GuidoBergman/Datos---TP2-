```python
import pandas as pd
```

```python
probas1 = pd.read_csv('predict_probas/knn_sin_outliers43.csv', index_col='building_id')
probas2 = pd.read_csv('predict_probas/xgb_baseline_001.csv', index_col='building_id')
probas3 = pd.read_csv('predict_probas/xgb_short1_001.csv', index_col='building_id')
```

```python
peso1 = 1
peso2 = 3
peso3 = 2
```

```python
merged = probas1.merge(probas2, right_index=True, left_index=True).merge(probas3, right_index=True, left_index=True)
```

```python
merged
```

```python
merged[1] = merged.proba_grado1_x * peso1 + merged.proba_grado1_y * peso2 + merged.proba_grado1 * peso3
merged[2] = merged.proba_grado2_x * peso1 + merged.proba_grado2_y * peso2 + merged.proba_grado2 * peso3
merged[3] = merged.proba_grado3_x * peso1 + merged.proba_grado3_y * peso2 + merged.proba_grado3 * peso3
```

```python
merged[[1,2,3]]
```

```python
prediction = merged[[1,2,3]].idxmax(axis="columns")
```

```python
submission_format = pd.read_csv('submission_format.csv', index_col='building_id')
my_submission = pd.DataFrame(data=prediction,
                             columns=submission_format.columns,
                             index=submission_format.index)
```

```python
my_submission.to_csv('subs/ensable_rankeado_001.csv')
```
