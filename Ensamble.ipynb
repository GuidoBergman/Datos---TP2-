```python
import pandas as pd
```

```python
probasA = pd.read_csv('predict_probas/xgb_baseline.csv', index_col='building_id')
probasB = pd.read_csv('predict_probas/xgb_short1_ambos_geos.csv', index_col='building_id')
probasC = pd.read_csv('predict_probas/lightGBM.csv', index_col='building_id')
probasD = pd.read_csv('predict_probas/catboost_2000_iterations.csv', index_col='building_id')
probasE = pd.read_csv('predict_probas/knn_sin_outliers43.csv', index_col='building_id')
```

```python
 {"peso_baseline": 4.219716801624352, "peso_lgbm": 2.39216802199134, "peso_short1": 3.411887326897567},
```

```python
pesoA = 4.219716801624352
pesoB = 3.411887326897567
pesoC = 2.39216802199134
pesoD = 1
pesoE = 0
```

```python
merged = probasA.merge(probasB, right_index=True, left_index=True)\
        .merge(probasC, right_index=True, left_index=True)\
        .merge(probasD, right_index=True, left_index=True)\
        .merge(probasE, right_index=True, left_index=True)
```

```python
merged
```

```python
merged[1] = merged.A1 * pesoA + merged.B1 * pesoB + merged.C1 * pesoC + merged.D1 * pesoD + merged.E1 * pesoE
merged[2] = merged.A2 * pesoA + merged.B2 * pesoB + merged.C2 * pesoC + merged.D2 * pesoD + merged.E2 * pesoE
merged[3] = merged.A3 * pesoA + merged.B3 * pesoB + merged.C3 * pesoC + merged.D3 * pesoD + merged.E3 * pesoE
```

```python
merged[[1,2,3]]
```

```python
prediction = merged[[1,2,3]].idxmax(axis="columns")
```

```python
submission_format = pd.read_csv('submission_format.csv', index_col='building_id')
my_submission = pd.DataFrame(data=prediction,
                             columns=submission_format.columns,
                             index=submission_format.index)
```

```python
my_submission.to_csv('subs/ensable_encontrador_de_pesos_001.csv')
```
